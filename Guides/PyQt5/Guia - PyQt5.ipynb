# Codigo basico

````python
from PyQt5 import QtWidgets
from PyQt5.QtWidgets import QApplication, QMainWindow
import sys


# Tomando todas las caracteristicas de QMainWindow
class MyWindow(QMainWindow):
    def __init__(self):
        super(MyWindow, self).__init__()
        self.setGeometry(760, 340, 300, 300)      # Xpos, Ypos, width, height
        self.setWindowTitle("Mi primera app")
        self.initUI()

    # Lo que va dentro de la app va aca
    def initUI(self):
        # Texto en la app
        self.label = QtWidgets.QLabel(self)
        self.label.setText("Esta es la label")
        self.label.move(50,50)

        # Botones
        self.b1 = QtWidgets.QPushButton(self)
        self.b1.setText("Boton")
        self.b1.move(150,100)
        self.b1.clicked.connect(self.oprimida)

    # Funcion para el boton
    def oprimida(self):
        self.label.setText("Papi, oprimiste el boton")
        # Este texto puede crear errores
        # Su tamaño es mas grande que el texto original por lo cual se va a cortar
        # Afecta a la ventana en la cual esta el texto

        # Actualizar automaticamente el tamaño del label para que quepa el nuevo texto
        self.label.adjustSize()

def window():
    app = QApplication(sys.argv)
    win = MyWindow()

    win.show()      # Mostrar la app
    sys.exit(app.exec())    # Cerrar app



window()
````
